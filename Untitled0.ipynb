# Data Transformation

I have decided to replace all occurrences of “<0.1” with 0.05:
- This is the mean of 0 and 0.1. Correctly formatted rational numbers are required in cases or regression models, meaning that the “<” would invalidate these values.

```python
df = df.replace("<0.1", '0.05')
```

I will replace the gradespan of 712 with 0 and 912 1 to more easily identify these variables as categorical values and not numerical values.

```python
df['GRADESPAN'] = df['GRADESPAN'].replace(712, 0)
df['GRADESPAN'] = df['GRADESPAN'].replace(912, 1)
```
